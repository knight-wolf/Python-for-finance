```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')
fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%B---%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator())

ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator())

ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location or Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))

ax.xaxis.set_minor_formatter(dates.DateFormatter('%a'))

fig.autofmt_xdate()
plt.tight_layout()
```
